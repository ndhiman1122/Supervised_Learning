```python
import numpy as np
import pandas as pd
from pathlib import Path

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
loans2019_df = pd.read_csv(Path('Resources/2019loans.csv'))
loans2020Q1_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
loans2019_df = pd.read_csv(Path('Resources/2019loans.csv'))
loans2020Q1_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training data

convertloans2019 = loans2019_df.drop(['loan_status', 'index'], axis=1)
loans2019 = pd.get_dummies(convertloans2019)
loans2019

convertloans2019 = loans2019_df['loan_status']
```

```python
# Convert categorical data to numeric and separate target feature for testing data

convertloans2020Q1 = loans2020Q1_df.drop(['loan_status', 'index'], axis=1)
loans2020Q1 = pd.get_dummies(convertloans2020Q1)
loans2020Q1

convertloans2020Q1 = loans2020Q1_df['loan_status']
```

```python
# add missing dummy variables to testing set

loans2020Q1['debt_settlement_flag_Y'] = 0
loans2020Q1
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score

classifier = LogisticRegression()

classifier.fit(loans2019, convertloans2019)
classifier.score(loans2020Q1, convertloans2020Q1)
                 
print(f"Training Data Score: {classifier.score(loans2019, convertloans2019)}")
print(f"Testing Data Score: {classifier.score(loans2020Q1, convertloans2020Q1)}")
```

```python
# Train a Random Forest Classifier model and print the model score

rfc_classifier = RandomForestClassifier()

rfc_classifier.fit(loans2019, convertloans2019)
rfc_classifier.score(loans2020Q1, convertloans2020Q1)
                 
print(f"Training Data Score: {rfc_classifier.score(loans2019, convertloans2019)}")
print(f"Testing Data Score: {rfc_classifier.score(loans2020Q1, convertloans2020Q1)}")
```

```python
# Scale the data

x_scaler = StandardScaler().fit(loans2019)
X_train_scaled = x_scaler.transform(loans2019)
X_test_scaled = x_scaler.transform(loans2020Q1)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score

scaled_classifier = LogisticRegression()
scaled_classifier.fit(X_train_scaled, convertloans2019)
scaled_classifier.score(X_test_scaled, convertloans2020Q1)
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score

scaled_rfc_classifier = RandomForestClassifier()

scaled_rfc_classifier.fit(X_train_scaled, convertloans2019)
scaled_rfc_classifier.score(X_test_scaled, convertloans2020Q1)
                 
print(f"Scaled Training Data Score: {scaled_rfc_classifier.score(X_train_scaled, convertloans2019)}")
print(f"Scaled Testing Data Score: {scaled_rfc_classifier.score(X_test_scaled, convertloans2020Q1)}")
```
